```python
import tensorflow as tf
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('../../sample_data/iris.csv')
```

```python
dataset.head(3)
```

```python
#Class nominal to index (0,1,2,..비어 있는 문자 없어야 함.)
dataset.loc[dataset.Species=='Iris-setosa', 'Species'] = 0
dataset.loc[dataset.Species=='Iris-versicolor', 'Species'] = 1
dataset.loc[dataset.Species=='Iris-virginica', 'Species'] = 2
dataset.head(3)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Dropout(0.1),
  tf.keras.layers.Dense(10, activation='relu'),
  tf.keras.layers.Dropout(0.1),
  tf.keras.layers.Dense(3, activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
X = dataset[dataset.columns[1:5]].values
Y = dataset.Species.values.astype(float)
train_X, test_X, train_y, test_y = train_test_split(X,Y, test_size=0.2)

# npX=df.loc[:,['F1','F2','F3','F4']].to_numpy()
# npY=df['Label'].to_numpy().reshape(-1,1)
# X_train, X_test, y_train, y_test = train_test_split(npX, npY, test_size=0.2)
```

```python
model.fit(train_X, train_y, epochs=500)

model.evaluate(test_X,  test_y, verbose=2)
```
